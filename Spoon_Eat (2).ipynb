## 2.2. Categorical Data

```python
def categ(data,column):
    # Define a color palette with 6 colors
    colors = sns.color_palette("husl", 6)

    # Create the figure and subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Plot the countplot in the left subplot
    sns.countplot(x=column, data=data, ax=ax1, palette=colors)
    ax1.set_title(column.upper(), fontsize=20, fontweight="bold")
    

    # Plot the pie chart in the right subplot
    wedges, texts, autotexts = ax2.pie(data[column].value_counts(), 
                                       labels=list(data[column].value_counts().index),
                                       autopct='%1.0f%%', 
                                       colors=colors,
                                       wedgeprops={"linewidth": 1, "edgecolor": "white"})
    ax2.axis("equal")
    ax2.legend(wedges, list(df[column].value_counts().index), 
               loc="center right", bbox_to_anchor=(1.2, 0.5))
    ax2.set_title(column.upper(), fontsize=20, fontweight="bold")

    # Set the font size of the pie chart labels
    for text in texts:
        text.set_fontsize(15)
    for autotext in autotexts:
        autotext.set_fontsize(15)
    
    plt.legend(loc="best")

    # Show the plot
    plt.show()
```

## 2.3. Correlation between Variabels

It indicates the degree to which changes in one variable are associated with changes in another variable. Correlation is often measured by a correlation coefficient, which is a value that ranges between -1 and 1. A positive correlation coefficient means that as one variable increases, the other variable also tends to increase, while a negative correlation coefficient means that as one variable increases, the other variable tends to decrease.

Scale of Correlation Coeficient:

0<=r<=0.19---Very Low Correlation

0.2<=r<=0.59---Low Correlation

0.4<=r<=0.59 Moderate Correlation

0.6<=r<=0.79---High Correlation

0.8<=r<=1.0---Very High Correlation

### 3.1.2. PCA

PCA stands for Principal Component Analysis. It is a dimensionality reduction technique used to transform high-dimensional datasets into a lower-dimensional space. It does so by identifying the principal components in the data, which are new variables that are linear combinations of the original features. These new variables are ordered in terms of how much variance they explain in the data.

Beacause in this case we have a lot of variabels we nee to apply PCA.

```python
from sklearn.decomposition import PCA
```

```python
# LetÂ´s set that we want 95% of variance in our data
pca = PCA(0.95)
df_cluster = df3.copy()
pca_result = pca.fit_transform(df_cluster)
```

```python
pca_result.shape
```

```python
df3.shape
```

```python
print ('Cumulative variance explained by 2 principal components: {:.2%}'.format(np.sum(pca.explained_variance_ratio_)))
```

```python
pca_result = pd.DataFrame(pca_result)
pca_result
```

### 3.1.3. Determine the K

```python
pca_result.info()
```

K is the number of clusters

__Elbow Method__

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,8):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=7,random_state=0)
    kmeans.fit(pca_result)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,8),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

__Dendogram__

```python
import scipy.cluster.hierarchy as sch
from matplotlib import pyplot
pyplot.figure(figsize=(12, 5))
dendrogram = sch.dendrogram(sch.linkage(pca_result, method = 'ward'))
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
plt.show()
```

By the dendogram and the Elbow Method the K appears to be 3

### 3.1.4. K-Means Cluster

```python
df_kmeans = pca_result.copy()
```

```python
# Training the K-Mean model
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
y_kmeans = kmeans.fit_predict(df_kmeans)
```

```python
df_kmeans
```

```python
# We called the df, that's why we need to refer to previous df to add cluster numbers
df_kmeans = df2.copy()
# Checking number of items in clusters and creating 'Cluster' column
df_kmeans['Cluster'] = y_kmeans
df_kmeans['Cluster'].value_counts()
```
